```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread('imagens/px-people.jpg')
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem);
```

```python
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
plt.imshow(imagem_gray, cmap="gray");
```

```python
classificador = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classificador.detectMultiScale(imagem_gray, 1.3, 5)
```

```python
len(faces)
```

```python
faces[0]
```

```python
imagem_anotada = imagem.copy()
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(imagem_anotada, (x, y), (x+w, y+h), (20, 255, 60), 2)
    
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(imagem_anotada);
```

```python
face_imagem = 0
for (x, y, w, h) in faces:
    face_imagem += 1
    imagem_roi = imagem[y:y+h, x:x+w]
    imagem_roi = cv2.cvtColor(imagem_roi, cv2.COLOR_BGR2RGB)
    cv2.imwrite(f"face_{face_imagem}.jpg", imagem_roi)
```
